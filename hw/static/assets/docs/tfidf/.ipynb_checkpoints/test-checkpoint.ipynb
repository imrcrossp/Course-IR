```python
def calc_idfmtx(func4idf, idf_count, tw, ts):
    mtx = [0]*tw
    for idx, freq in enumerate(idf_count['_idf']):
        mtx[idx] = abs(func4idf(ts, freq))
    return mtx
```

```python
def text_prepare(strr):
    import re
    strr = strr.lower()
    REPLACE_BY_SPACE_RE = re.compile('[/(){}\[\]\|@,;]')
    BAD_SYMBOLS_RE = re.compile('[^a-z\ ]')
    STOPWORDS = set(nltk.corpus.stopwords.words('english'))

    strr = REPLACE_BY_SPACE_RE.sub(" ", strr)
    strr = BAD_SYMBOLS_RE.sub("", strr)
    sent = " ".join([w for w in nltk.word_tokenize(strr) if w not in STOPWORDS]) 
    return sent
```

```python
def calc_tf(func4tf, keys, remtx, text, tw):
    dyfreq = [0]*tw 
    text = text_prepare(text)
    for _old, new_ in zip(remtx['old'], remtx['new']):
        dystr = text.replace(_old, new_)
    dytoken = nltk.word_tokenize(dystr)
    for idx, i in enumerate(dytoken):
        for _old, new_ in zip(remtx['old'], remtx['new']):
            if i == new_:
                dytoken[idx] = _old
    for i in dytoken:
        tmp = keys.index[(keys['keys'] == i)].tolist()
        if len(tmp) == 1:
            dyfreq[tmp[0]] = dyfreq[tmp[0]]+1
        else:
            print(i)
    for i in range(tw):
        dyfreq[i] = func4tf(dyfreq[i], len(dytoken))
    return dyfreq
```

```python
text = "The hepatitis B vaccine is a medication used in the prevention of hepatitis B infection, which can lead to chronic liver failure and hepatocellular carcinoma."
def sort_cossim(text):
    from numpy import dot
    from numpy.linalg import norm
    result = {}
    tf_nat = lambda x, y: x/y
    tf_log = lambda x, y: math.log(1+x)
    tf_aug = lambda x, y: 0.5+0.5*((x)/(y))
    '''------------------------------------'''
    idf_nat = lambda x, y: math.log(x/(1+y))
    idf_smt = lambda x, y: math.log(x/(1+y))+1
    idf_pro = lambda x, y: math.log((x-y)/(1+y))
    tf = pd.read_csv("./tf.csv")
    dfidf_count = pd.read_csv("./idf_count.csv") 
    dfremtx = pd.read_csv("./dict.csv")
    dfkeys = pd.read_csv("./keys.csv")
    dfbook = pd.read_csv("./book.csv")
    ts = int(tf['freq'][1])
    tw = int(tf['freq'][2])
    idfmtx = calc_idfmtx(idf_nat, dfidf_count, tw, ts)
    Amtx = calc_tf(tf_nat, dfkeys, dfremtx, text, tw)
    Awei = np.multiply(Amtx, idfmtx)
    for i in range(ts):
        sent, ori = dfbook['processed'][i], dfbook['nonprocessed'][i]
        Bmtx = calc_tf(tf_nat, dfkeys, dfremtx, sent, tw)
        Bwei = np.multiply(Bmtx, idfmtx)
        angle = dot(Awei, Bwei)/(norm(Awei)*norm(Bwei))
        result[i] = angle
    result = sorted(result.items(), key=lambda x:x[1], reverse = True)
    return result
```

```python
ans = sort_cossim(text)
```
